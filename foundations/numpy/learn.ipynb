```python
def dot_product(l1,l2):
      def check(l1,l2):
            # if it's 1-D
            if type(l1[0])==list and type(l2[0] == list):
                  if len(l1[0])==len(l2):
                              return True
                  else:
                              return "dimension not compatible!"
            else:
                  if len(l1)!=len(l2):
                        raise ValueError("dimension not compatible!")
                  else:
                        return False
      def dot_prod(l1,l2):
            list_dot = 0
            # for i in range(len(l1)):
            #       list_dot+=(l1[i]*l2[i])
            return sum([x*y for x,y in zip(l1,l2)])
            # return list_dot
      
      # START of dot prod
      final = []
      if len(l1) == 0 or len(l2) == 0:
            raise ValueError("empty list not allowed sarr.")

      if check(l1,l2)==True: #it's 2d array
            for i in range(len(l1)):
                  
                  lst = []
                  k =val = 0
                  while k<len(l2[0]):
                        temp = []
                        for j in range(len(l2)):
                              # print(j,k)
                              temp.append(l2[j][k])
                        k+=1
                        lst.append(temp)
                  temp=[]
                  for l in lst:
                        # print(l1[i], l)
                        temp.append(dot_prod(l1[i], l))
                  
                  final.append(temp)
            return final
      elif check(l1,l2)==False: #means it's 1-D
            return dot_prod(l1,l2)
      else:
            return check(l1,l2)
      
```

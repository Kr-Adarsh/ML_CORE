# Random Creations

```python
# Define matrix dimensions
m = 5
n = 5

print("### 1. Matrix with Random Integers")
# Method 1: Using np.random.randint()
df_rand_int = pd.DataFrame(np.random.randint(0, 100, size=(m, n)))
print("DataFrame with random integers (0-99):")
display(df_rand_int)

print("\n### 2. Matrix with Random Floating-Point Numbers")
# Method 2: Using np.random.rand() (uniform distribution [0, 1))
df_rand_float_uniform = pd.DataFrame(np.random.rand(m, n))
print("DataFrame with random floats (uniform [0, 1)):")
display(df_rand_float_uniform)

# Method 3: Using np.random.randn() (standard normal distribution)
df_rand_float_normal = pd.DataFrame(np.random.randn(m, n))
print("DataFrame with random floats (standard normal distribution):")
display(df_rand_float_normal)

print("\n### 3. Matrix with Sequential Numbers")
# Method 4: Using np.arange()
df_sequential = pd.DataFrame(np.arange(m * n).reshape(m, n))
print(f"DataFrame with sequential numbers (0 to {m*n-1}):")
display(df_sequential)

# You can also customize column and index names
df_custom = pd.DataFrame(
    np.random.randint(0, 10, size=(m, n)),
    columns=[f'Col_{i}' for i in range(n)],
    index=[f'Row_{i}' for i in range(m)]
)
print("\nDataFrame with custom column and index names:")
display(df_custom)
```

## Summary of Differences

| Feature           | `.loc`                          | `.iloc`                           |
| :---------------- | :------------------------------ | :-------------------------------- |
| **Indexing Type** | Label-based                     | Integer-position based            |
| **Input**         | Row/Column labels, boolean array| Integer positions, boolean array  |
| **Slicing End**   | Inclusive                       | Exclusive                         |
| **Usage**         | When you know the names/labels  | When you know the positions       |

```python
display(df.loc["2020-01-01":"2020-01-04", "A":"C"])
display(df.loc["2020-01-01":"2020-01-04", ["A","C"]])

display(df.iloc[0:4, 0:3])
display(df.iloc[0:4, [0,3]])
```

```python
#@title grouping - GroupBy

df = pd.DataFrame(
    {
        "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo"],
        "B": ["one", "one", "two", "three", "two", "two", "one", "three"],
        "C": np.random.randn(8),
        "D": np.random.randn(8),
    }
)

df
```

```python
df.groupby("A")[["C", "D"]].sum() #this is giving sum of all vals in C and D corresponding to row A uniques..
```

```python
display(df)
df.groupby(['A','B']).sum() # original sum operation

print("\n--- Mean of C and D by A and B ---")
display(df.groupby(['A','B'])[['C','D']].mean())

print("\n--- Minimum of C and D by A and B ---")
display(df.groupby(['A','B'])[['C','D']].min())

print("\n--- Maximum of C and D by A and B ---")
display(df.groupby(['A','B'])[['C','D']].max())

print("\n--- Count of C and D by A and B ---")
display(df.groupby(["A", "B"])[['C','D']].count())
```

```python
#@title Categoricals data analysis
```

```python
df = pd.DataFrame(
    {"id": [1,2,3,4,5,6], "raw_grade": ["a", "b", "b", "a","e", "a"]}
)
df
```

```python
# Converting the raw grades to a categorical data type:

df["grade"] = df["raw_grade"].astype("category")
df["grade"]
```

```python
# Rename the categories to more meaningful name
# df["grade"].cat.rename_categories = ["very good", "good", "very bad"]
# df["grade"]

new_category = ["very_good", "good", " bad"]
df["grade"] = df["grade"].cat.rename_categories(new_category)
df["grade"]
```

```python
df.sort_values(by="grade")
df.groupby("grade", observed= False).size()
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv("train.csv")
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from tqdm.auto import tqdm
nltk.download('punkt_tab')
```

```python
df[df['post_id']==73]['comment']
```

```python
# tokaniser:

text = df['comment'][73]
tokens = nltk.word_tokenize(text)
```

```python
# to get the part of speech
nltk.download('averaged_perceptron_tagger_eng')
tagged = nltk.pos_tag(tokens) #learn about it on nltk pos tagging
# Some examples: NN=Noun, JJ=Adjective, VB=Verb, PRP=Pronoun 
```

```python
nltk.download('words')
nltk.download('maxent_ne_chunker_tab')
entities = nltk.chunk.ne_chunk(tagged)
# print(entities)
entities.pprint() #pretty print
```

```python
#sentiment score
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()
comments_list = df['comment'].astype(str).tolist()

print("Running lightning-fast VADER Sentiment Analysis...")
sentiment_scores = [sia.polarity_scores(text)['compound'] for text in tqdm(comments_list)]
df['sentiment_vader'] = sentiment_scores
print("Done..")
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from tqdm.auto import tqdm
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()
```

```python
res = {}
for i, row in tqdm(df[:10000].iterrows(), total=len(df[:10000])):
    text = row['comment']
    myid = row['post_id']
    res[myid] = sia.polarity_scores(text)
```

```python
# res
vaders = pd.DataFrame(res).T
vaders.reset_index().rename(columns={'index': 'post_id'})
vaders
```

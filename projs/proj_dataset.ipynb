```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("customer_shopping_behavior.csv")
df.head()
```

```python
# df['Review Rating'] = df['Review Rating'].fillna(df["Category"].median())
df["Review Rating"] = df.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# it's good to write everything in lowercase and napcases
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", "_")
df.columns
# we found that "'purchase_amount_(usd)'" looks urregular so let's manually change it
df = df.rename(columns={"purchase_amount_(usd)": "purchase_amount"})
df.columns
# now it's neat and consistent
```

```python
# let's do feature engineering now :
# create a column age_group
```

```python
label = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q =4, labels= label)
```

```python
df[['age', 'age_group']].head(10)
```

```python
# let's create another columns purchase_frequency_days
frequence_mapping = {
    "Fortnightly":14,
    "Weekly": 7,
    "Monthly": 30,
    "Quarterly": 90,
    "Bi-Weekly": 14,
    "Annually": 365,
    "Every 3 Months": 90
}

df["purchase_frequency_days"] = df["frequency_of_purchases"].map(frequence_mapping)
```

```python
df[["frequency_of_purchases","purchase_frequency_days"]].head(10)
```

```python
df[["discount_applied", "promo_code_used"]].head(10)
# Q: do we actually need both of these columsn?? as we both of equal weitage:?
# cehck is both are same?
(df["discount_applied"] == df['promo_code_used']).all()
```

```python
# we got True thus we can remove one of them
df = df.drop("promo_code_used", axis = 1)
```

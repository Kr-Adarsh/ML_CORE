### Least Square Approximation

```python
np.random.seed(100)
x = np.linspace(1,3, 5)
y = np.linspace(1,3, 5)


# now we need A [[x1, 1], [x2,1], ...]
A = np.vstack([x,np.ones(len(x))]).T
b = y

# we know lsa= A^T*A*X^ = A^T*b or
            # X^ = (A^T*A)^-1 * A^T*b
def lsa(A,b):
    At = A.T
    At_A = np.dot(At, A)
    At_A_inverse = np.linalg.inv(At_A)
    Atb = np.dot(At, b)

    # finally 
    x_ = At_A_inverse @ Atb
    slope, intercept = x_[0], x_[1]
    return slope, intercept

print(x, y)
print(lsa(A, b))
```
